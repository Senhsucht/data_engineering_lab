# Ejercicio 2: Agregación de Datos

Ahora que dominas la limpieza básica, subiremos un poco el nivel.

## Escenario:

La gerencia quiere entender las ventas por producto. Necesitan saber:
1. El precio promedio de cada producto.
2. El número total de ventas (conteo de registros) de cada producto.
3. La fecha de la primera y la última venta para cada producto.

## Instrucciones:

Utiliza el DataFrame limpio que creaste (ventas_limpias.csv) y pandas para:
1. Agrupar los datos por la columna producto.
2. Calcular agregaciones: Dentro de cada grupo, calcula el precio promedio, el conteo total de ventas y la fecha mínima (min) y máxima (max).
3. Guarda el resultado: Guarda el DataFrame resultante en un nuevo archivo llamado ventas_agregadas.csv.

Generado por: Gemini 2.5 Flash

```python
import pandas as pd
```

```python
input_file = '../outputs/ventas_limpias.csv'
output_file = '../outputs/ventas_agregadas.csv'
```

```python
df = pd.read_csv(input_file)
df
```

```python
# Se agrupa por producto y se agregan las columnas de precio promedio, ventas totales, fecha mínima y máxima.
# Se ordena por producto de forma ascendente.
# Nota: La columna producto se vuelve el indice de la tabla, para poder usarse como columna se usa reset_index()
df = df.groupby(['producto']).agg(
    precio_promedio=('precio','mean'),
    ventas_totales=('producto','count'),
    fecha_min=('fecha','min'),
    fecha_max=('fecha','max')
).sort_values(by='producto', ascending=True).reset_index()

df
```

```python
# Se guarda el dataframe en un csv.
df.to_csv(output_file, index=False)
```
